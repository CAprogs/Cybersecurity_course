Working with CSV Files

```python
import csv

data = [
    ["Name", "Age", "City"],
    ["Alice", 25, "New York"],
    ["Bob", 30, "San Francisco"],
]

# Writing to a CSV file
with open("data.csv", "w", newline="") as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerows(data)

# Reading from a CSV file
with open("data.csv", "r") as csv_file:
    csv_reader = csv.reader(csv_file)
    for row in csv_reader:
        print(row)
```

Regular Expressions (Regex)

```python
import re

text = "The price of the product is $20.50"

match = re.search(r'\$(\d+\.\d+)', text)
if match:
    price = float(match.group(1))
    print("Product Price:", price)
```

sqlite3 Module - SQLite Database

```python
import sqlite3

conn = sqlite3.connect("example.db")
cursor = conn.cursor()

cursor.execute("CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT, age INTEGER)")
conn.commit()

# Inserting data
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("Alice", 25))
conn.commit()

# Querying data
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()
print("Database Rows:", rows)

conn.close()
```

```python
import seaborn as sns

# Load the 'tips' dataset from seaborn
tips = sns.load_dataset('tips')

# Display the first few rows of the dataset
print(tips.head())
```

Descriptive Statistics

```python
# Display summary statistics of the numeric columns
print(tips.describe())
```

 Data Visualization - Pair Plot

```python
import matplotlib.pyplot as plt

# Visualize pairwise relationships in the dataset
sns.pairplot(tips, hue='sex', markers=['o', 's'], height=3)
plt.show()
```

Data Visualization - Box Plot

```python
# Create a box plot to visualize the distribution of total bill amounts
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset from seaborn
tips = sns.load_dataset('tips')

# Select only numeric columns for correlation analysis
numeric_columns = tips.select_dtypes(include=['float64', 'int64'])

# Display a heatmap of the correlation matrix
correlation_matrix = numeric_columns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

Categorical Plot - Bar Chart

```python
# Create a bar chart to show the average total bill for each day
sns.barplot(x='day', y='total_bill', data=tips, ci=None)
plt.show()
```

Distribution Plot

```python
# Display the distribution of total bill amounts using a histogram
sns.histplot(tips['total_bill'], kde=True, bins=20)
plt.show()
```

Violin Plot

```python
# Create a violin plot to show the distribution of total bill amounts for each day
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

Regression Plot

```python
# Visualize the relationship between total bill and tip with a regression line
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()
```

Count Plot

```python
# Create a count plot to show the number of observations for each day
sns.countplot(x='day', data=tips)
plt.show()
```
